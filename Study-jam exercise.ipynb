7. What those unique values are and counts for each

```python
df.columns
```

```python
for feature in df.columns[:]:
    df[feature].dropna(inplace=True)#drop the null values in each column
    print('***' + feature + '***')
    display(df[feature].value_counts())#get the values and counts of non-null unique values
    print()
```

```python
for column in df.columns:
    print(f"The following are the counts of each unique value in the {column} column:")
    print(df[str(column)].value_counts())
    print('\n')
```
